```python
# let's take a look at a transaction
# it is not very readable
tx = w3.eth.get_transaction('0x693ccb83ccad308f09f6158206d6950bf0d0c3ed0a4fd4769c47caa49da60c54')
tx
```

```python
# this is the input data passed to the smart contract that was called
tx['input']
```

```python
# function calls are the keccak hash of the function name and the types of the parameters
Web3.keccak(text='transfer(address,uint256)').hex()[0:10]
```

```python

# in this case, the function is called 'transfer' and it takes two parameters, an address and an integer
# the address is address, and the value uint256
# the first 4 bytes are the function hash, and the rest are the parameters


function_hash = tx['input'][0:10]
address = tx['input'][34:74]
value = tx['input'][74:]

print(f"function_hash: {function_hash}")
print(f"address: {address}")
print(f"value: {w3.toInt(hexstr=value)/10**decimals}")
```
